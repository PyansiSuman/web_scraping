```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
website_url = requests.get('https://books.toscrape.com/').text
soup = BeautifulSoup(website_url,'lxml')
```

```python
website_url
```

```python
all_div = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
len(all_div)
```

```python
img = []
rating = []
avability = []
price = []
name = []
for i in all_div:
    img.append(i.find('img', class_='thumbnail').get('src'))
    avability.append(i.find('p', class_='instock availability').text.strip())
    price.append(i.find('p', class_='price_color').text.strip())
    name.append(i.find('h3').text.strip())
    rating.append(i.find('p').get('class'))

```

```python
d = {'name':name,'price':price,'avability':avability,'rating':rating,'img':img}
df = pd.DataFrame(d)
```
